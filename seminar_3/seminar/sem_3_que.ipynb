# Mathematical Pendulum Equation

Let's consider the equation of a mathematical pendulum without friction, where all constants are set to one.

## Basic Equation

The standard equation of pendulum motion has the form:

$$\ddot{\theta} + \frac{g}{L}\sin(\theta) = 0$$

Where:
- $\theta$ - angle of deviation from vertical
- $g$ - acceleration due to gravity
- $L$ - pendulum length

## Simplified Equation

If we set $g = 1$ and $L = 1$, we get a simplified equation:

$$\ddot{\theta} + \sin(\theta) = 0$$

This is a nonlinear second-order differential equation.

## Linear Approximation

For small oscillations (small angles $\theta$), we use the approximation $\sin(\theta) \approx \theta$, which gives a linear equation:

$$\ddot{\theta} + \theta = 0$$

This is a harmonic oscillator equation with the general solution:

$$\theta(t) = A\cos(t + \phi)$$

where $A$ is the amplitude of oscillations, and $\phi$ is the initial phase.

## Nonlinear Case

For large deviation angles, the equation remains nonlinear, and its exact analytical solution is expressed through elliptic integrals. The period of oscillations depends on the amplitude and can be calculated using the formula:

$$T = 4\int_0^{\theta_0} \frac{d\theta}{\sqrt{2(\cos\theta - \cos\theta_0)}}$$

where $\theta_0$ is the amplitude of oscillations.

# Mathematical Pendulum Equation with Linear Control

Let's add linear control to the pendulum equation to study its dynamics under external influence.

## Basic Equation with Control

The original pendulum equation without friction, with the addition of control action, takes the form:

$\ddot{\theta} + \sin(\theta) = a(t)$

where $a(t)$ is the control function, depending on time.

## Linear Feedback Control

The most common form of linear control is a linear combination of state variables:

$a(t) = k_1 \cdot \theta + k_2 \cdot \dot{\theta}$

where $k_1$ and $k_2$ are feedback coefficients.

Substituting this expression into the pendulum equation, we get:

$\ddot{\theta} + \sin(\theta) = k_1 \cdot \theta + k_2 \cdot \dot{\theta}$

or

$\ddot{\theta} + \sin(\theta) - k_1 \cdot \theta - k_2 \cdot \dot{\theta} = 0$

## Linear Approximation with Control

For small deviation angles, using $\sin(\theta) \approx \theta$, we get:

$\ddot{\theta} + \theta = a(t)$

And considering linear feedback control:

$\ddot{\theta} + \theta - k_1 \cdot \theta - k_2 \cdot \dot{\theta} = 0$

Or in regrouped form:

$\ddot{\theta} + (1 - k_1) \cdot \theta - k_2 \cdot \dot{\theta} = 0$

Proper selection of coefficients $k_1$ and $k_2$ allows changing the dynamic characteristics of the system, including its stability and the rate of oscillation decay.

## Matrix Form Representation

We can convert the second-order differential equation to a system of first-order equations by introducing a state vector. Let's define the state vector s:

$s_1 = \theta$
$s_2 = \dot{\theta}$

### For the nonlinear case with control

$\dot{s}_1 = s_2$
$\dot{s}_2 = -\sin(s_1) + a(t)$

In matrix form, this can be written as:

$\begin{bmatrix} \dot{s}_1 \\ \dot{s}_2 \end{bmatrix} = \begin{bmatrix} 0 & 1 \\ 0 & 0 \end{bmatrix} \begin{bmatrix} s_1 \\ s_2 \end{bmatrix} + \begin{bmatrix} 0 \\ -\sin(s_1) \end{bmatrix} + \begin{bmatrix} 0 \\ 1 \end{bmatrix} a(t)$

Or, more compactly, as a nonlinear state-space model:

$\dot{\mathbf{s}} = f(\mathbf{s}) + g \cdot a(t)$

Where:
$\mathbf{s} = \begin{bmatrix} s_1 \\ s_2 \end{bmatrix}, \quad f(\mathbf{s}) = \begin{bmatrix} s_2 \\ -\sin(s_1) \end{bmatrix}, \quad g = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$

### For the linearized case with control

For small oscillations, where $\sin(\theta) \approx \theta$ or equivalently $\sin(s_1) \approx s_1$:

$\dot{s}_1 = s_2$
$\dot{s}_2 = -s_1 + a(t)$

In standard state-space form:

$\begin{bmatrix} \dot{s}_1 \\ \dot{s}_2 \end{bmatrix} = \begin{bmatrix} 0 & 1 \\ -1 & 0 \end{bmatrix} \begin{bmatrix} s_1 \\ s_2 \end{bmatrix} + \begin{bmatrix} 0 \\ 1 \end{bmatrix} a(t)$

### For the linearized case with feedback control

With the feedback control law $a(t) = k_1s_1 + k_2s_2$, the closed-loop system becomes:

$\begin{bmatrix} \dot{s}_1 \\ \dot{s}_2 \end{bmatrix} = \begin{bmatrix} 0 & 1 \\ -1+k_1 & k_2 \end{bmatrix} \begin{bmatrix} s_1 \\ s_2 \end{bmatrix}$

This matrix representation is useful for analyzing the system's stability and control properties using eigenvalue techniques from linear systems theory.

# Analysis of Linear Dynamical Systems: Eigenvalues, Determinant, and Trace

## Introduction

This document provides rules and interpretations for analyzing linear dynamical systems using eigenvalues, determinant, and trace of the system matrix. These properties determine the qualitative behavior of a system near its equilibrium points.

For a linear system represented in the form:

$\dot{\mathbf{s}} = A\mathbf{s}$

where $\mathbf{s} \in \mathbb{R}^n$ is the state vector and $A$ is the $n \times n$ system matrix, the qualitative behavior of trajectories near the origin is determined by the eigenvalues of $A$.

## Eigenvalues

Eigenvalues of the system matrix $A$ provide critical information about the stability and dynamics of the system.

### Definition
The eigenvalues $\lambda_i$ of matrix $A$ are the solutions to the characteristic equation:

$$\det(A - \lambda I) = 0$$

### Interpretation Rules

#### Real Eigenvalues

1. **Negative Real Eigenvalues ($\lambda_i < 0$ for all $i$)**
   - System is asymptotically stable
   - Trajectories converge to the equilibrium point without oscillations
   - Convergence rate in the direction of corresponding eigenvector is $e^{\lambda_i t}$
   - Classified as a stable node when all eigenvalues are distinct

2. **Positive Real Eigenvalues ($\lambda_i > 0$ for at least one $i$)**
   - System is unstable
   - Trajectories diverge from the equilibrium point without oscillations
   - Divergence rate in the direction of corresponding eigenvector is $e^{\lambda_i t}$
   - Classified as an unstable node when all eigenvalues are positive

3. **Mixed Sign Real Eigenvalues**
   - System has a saddle point (always unstable)
   - Some trajectories approach the equilibrium along stable manifolds
   - Other trajectories move away along unstable manifolds
   - Direction of approach/divergence determined by corresponding eigenvectors

4. **Zero Eigenvalue ($\lambda_i = 0$ for some $i$)**
   - Non-isolated equilibrium point or structural instability
   - System has a continuum of equilibria in some direction
   - Linear analysis insufficient, requires nonlinear analysis
   - Often indicates a bifurcation point

5. **Repeated Real Eigenvalues**
   - If all negative: critically damped system (fastest non-oscillatory return)
   - May indicate a degenerate node (trajectories approach from one direction)

#### Complex Eigenvalues

For 2D systems, complex eigenvalues always appear in conjugate pairs: $\lambda = \alpha \pm i\beta$

1. **Negative Real Part ($\alpha < 0$)**
   - System is asymptotically stable
   - Trajectories spiral inward toward the equilibrium point
   - Classified as a stable focus (or spiral)
   - Damping rate determined by $\alpha$
   - Oscillation frequency determined by $\beta$

2. **Positive Real Part ($\alpha > 0$)**
   - System is unstable
   - Trajectories spiral outward away from the equilibrium point
   - Classified as an unstable focus (or spiral)
   - Growth rate determined by $\alpha$
   - Oscillation frequency determined by $\beta$

3. **Zero Real Part ($\alpha = 0$)**
   - System is neutrally stable (conservative)
   - Trajectories form closed orbits around the equilibrium point
   - Classified as a center
   - Oscillation frequency determined by $\beta$
   - No damping or amplification

## Determinant

The determinant of matrix $A$ provides information about volume changes in phase space and the nature of the equilibrium point.

### Definition
For a 2×2 matrix $A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}$:

$$\det(A) = ad - bc = \lambda_1 \lambda_2$$

For an $n \times n$ matrix, the determinant equals the product of all eigenvalues:

$$\det(A) = \prod_{i=1}^{n} \lambda_i$$

### Interpretation Rules

1. **Positive Determinant ($\det(A) > 0$)**
   - Eigenvalues have the same sign (both positive or both negative)
   - Equilibrium is either a node or a focus (stable or unstable)
   - For 2D systems:
     - If $\det(A) < 1$: Phase space volumes contract
     - If $\det(A) > 1$: Phase space volumes expand
     - If $\det(A) = 1$: Phase space volumes preserved (area-preserving map)

2. **Negative Determinant ($\det(A) < 0$)**
   - Eigenvalues have opposite signs
   - Equilibrium is a saddle point (always unstable)
   - Phase space is stretched in some directions and compressed in others

3. **Zero Determinant ($\det(A) = 0$)**
   - At least one eigenvalue is zero
   - Matrix is singular (non-invertible)
   - System has a lower-dimensional invariant subspace
   - Linear analysis insufficient, requires nonlinear analysis

## Trace

The trace of matrix $A$ provides information about divergence and average growth rates.

### Definition
For a matrix $A$, the trace is the sum of diagonal elements:

$$\text{tr}(A) = \sum_{i=1}^{n} a_{ii} = \sum_{i=1}^{n} \lambda_i$$

The trace equals the sum of all eigenvalues.

### Interpretation Rules

1. **Negative Trace ($\text{tr}(A) < 0$)**
   - Sum of eigenvalues is negative
   - On average, the system is contractive
   - Energy is dissipated (damping is present)
   - For 2D systems, may be stable if $\det(A) > 0$

2. **Positive Trace ($\text{tr}(A) > 0$)**
   - Sum of eigenvalues is positive
   - On average, the system is expansive
   - Energy is added to the system (anti-damping)
   - For 2D systems, at least one eigenvalue has positive real part (unstable)

3. **Zero Trace ($\text{tr}(A) = 0$)**
   - Sum of eigenvalues is zero
   - System is conservative (Hamiltonian)
   - Energy is conserved
   - Phase space volumes are preserved
   - 2D systems with zero trace and positive determinant have pure imaginary eigenvalues (centers)

## Combined Analysis for 2D Systems

For 2×2 systems, the characteristic equation is:

$$\lambda^2 - \text{tr}(A)\lambda + \det(A) = 0$$

The eigenvalues are:

$$\lambda_{1,2} = \frac{\text{tr}(A) \pm \sqrt{\text{tr}(A)^2 - 4\det(A)}}{2}$$

The discriminant $\Delta = \text{tr}(A)^2 - 4\det(A)$ determines the nature of eigenvalues:

1. **$\Delta < 0$**: Complex conjugate eigenvalues (focus or center)
2. **$\Delta = 0$**: Repeated real eigenvalues (degenerate node or critical damping)
3. **$\Delta > 0$**: Distinct real eigenvalues (node or saddle)

### Stability Regions in the Trace-Determinant Plane

The stability of a 2D system can be visualized in the trace-determinant plane:

- **Stable Region**: $\det(A) > 0$ and $\text{tr}(A) < 0$
  - Stable node: $\Delta > 0$
  - Stable focus: $\Delta < 0$
  - Critically damped: $\Delta = 0$

- **Unstable Region**: $\det(A) < 0$ or $\text{tr}(A) > 0$
  - Saddle point: $\det(A) < 0$
  - Unstable node: $\det(A) > 0$, $\text{tr}(A) > 0$, $\Delta > 0$
  - Unstable focus: $\det(A) > 0$, $\text{tr}(A) > 0$, $\Delta < 0$

- **Neutral Stability**: $\det(A) > 0$ and $\text{tr}(A) = 0$
  - Center: $\Delta < 0$ (pure imaginary eigenvalues)

## Application to Pendulum Systems with Control

For a linear pendulum system with control parameters $k_1$ and $k_2$, the system matrix is:

$$A = \begin{pmatrix} 0 & 1 \\ -1+k_1 & k_2 \end{pmatrix}$$

The determinant and trace are:
- $\det(A) = 1-k_1$
- $\text{tr}(A) = k_2$


## Reference Table: Equilibrium Types and Their Properties

| Equilibrium Type | Eigenvalues | Determinant | Trace | Discriminant | Stability | Behavior |
|------------------|-------------|-------------|-------|-------------|-----------|----------|
| Stable Node | $\lambda_1,\lambda_2 < 0$, real | $>0$ | $<0$ | $>0$ | Stable | Non-oscillatory convergence |
| Stable Focus | $\Re(\lambda_{1,2}) < 0$, complex | $>0$ | $<0$ | $<0$ | Stable | Damped oscillations |

| Degenerate Case | One $\lambda = 0$ | $=0$ | Any | Any | Special | Requires nonlinear analysis |

## Practical Guidelines for System Analysis

1. **Calculate the system matrix** $A$ from the governing equations
2. **Find determinant and trace** of the matrix
3. **Calculate the discriminant** $\Delta = \text{tr}(A)^2 - 4\det(A)$
4. **Determine eigenvalues** using the characteristic equation
5. **Classify equilibrium type** using the reference table
6. **Interpret physical behavior** based on eigenvalues

This approach provides a complete qualitative understanding of the system's behavior near equilibrium points without solving the differential equations explicitly.

# Analysis of Stability Regions for a Linear System

## Original System

For a linear pendulum system with control parameters $k_1$ and $k_2$, the system matrix is:

$$A = \begin{pmatrix} 0 & 1 \\ -1+k_1 & k_2 \end{pmatrix}$$

For this matrix:
- Determinant: $\det(A) = 1-k_1$
- Trace: $\text{tr}(A) = k_2$
- Discriminant: $\Delta = \text{tr}(A)^2 - 4\det(A) = k_2^2 - 4(1-k_1)$

## Boundary Regions in the $k_1$-$k_2$ Parameter Plane

1. **Determinant boundary**: $k_1 = 1$
   - $\det(A) > 0$: $k_1 < 1$
   - $\det(A) < 0$: $k_1 > 1$

2. **Trace boundary**: $k_2 = 0$
   - $\text{tr}(A) > 0$: $k_2 > 0$
   - $\text{tr}(A) < 0$: $k_2 < 0$

3. **Discriminant boundary**: $k_2^2 = 4(1-k_1)$
   - $\Delta > 0$: $k_2^2 > 4(1-k_1)$
   - $\Delta < 0$: $k_2^2 < 4(1-k_1)$

## Equilibrium Type Regions

1. **Stable Node**:
   - Conditions: $k_1 < 1$, $k_2 < 0$, $k_2^2 > 4(1-k_1)$
   - Eigenvalues: real negative ($\lambda_1, \lambda_2 < 0$)
   - Behavior: non-oscillatory convergence to equilibrium point

2. **Stable Focus**:
   - Conditions: $k_1 < 1$, $k_2 < 0$, $k_2^2 < 4(1-k_1)$
   - Eigenvalues: complex conjugate with negative real part
   - Behavior: damped oscillations

3. **Center**:
   - Conditions: $k_1 < 1$, $k_2 = 0$
   - Eigenvalues: purely imaginary ($\lambda_{1,2} = \pm i\omega$)
   - Behavior: undamped oscillations

4. **Unstable Focus**:
   - Conditions: $k_1 < 1$, $k_2 > 0$, $k_2^2 < 4(1-k_1)$
   - Eigenvalues: complex conjugate with positive real part
   - Behavior: growing oscillations

5. **Unstable Node**:
   - Conditions: $k_1 < 1$, $k_2 > 0$, $k_2^2 > 4(1-k_1)$
   - Eigenvalues: real positive ($\lambda_1, \lambda_2 > 0$)
   - Behavior: non-oscillatory divergence from equilibrium point

6. **Saddle Point**:
   - Conditions: $k_1 > 1$ (for any value of $k_2$)
   - Eigenvalues: real with opposite signs ($\lambda_1 < 0 < \lambda_2$)
   - Behavior: unstable, mixed

7. **Critical Damping**:
   - Conditions: $k_1 < 1$, $k_2 = -2\sqrt{1-k_1}$
   - Eigenvalues: repeated negative ($\lambda_1 = \lambda_2 < 0$)
   - Behavior: fastest non-oscillatory return

8. **Degenerate Case**:
   - Conditions: $k_1 = 1$ (for any value of $k_2$)
   - Eigenvalues: one eigenvalue equals zero
   - Behavior: requires nonlinear analysis

## Eigenvalue Equations

For this system, the characteristic equation is:

$$\lambda^2 - \text{tr}(A)\lambda + \det(A) = 0$$

$$\lambda^2 - k_2\lambda + (1-k_1) = 0$$

Eigenvalues:

$$\lambda_{1,2} = \frac{k_2 \pm \sqrt{k_2^2 - 4(1-k_1)}}{2}$$

# Stability Proof using Lyapunov Function

## Problem Statement

Consider the dynamical system given by:

$$\begin{bmatrix} \dot{s}_1 \\ \dot{s}_2 \end{bmatrix} = \begin{bmatrix} 0 & 1 \\ -1+k_1 & k_2 \end{bmatrix} \begin{bmatrix} s_1 \\ s_2 \end{bmatrix}$$

with parameters $(k_1, k_2) = (1/2, -1)$.

We need to prove the stability of this system using Lyapunov's direct method.

## Solution

### Step 1: Substitute the Parameter Values

First, let's substitute the parameter values into the system matrix:

$$\begin{bmatrix} \dot{s}_1 \\ \dot{s}_2 \end{bmatrix} = \begin{bmatrix} 0 & 1 \\ -1+1/2 & -1 \end{bmatrix} \begin{bmatrix} s_1 \\ s_2 \end{bmatrix} = \begin{bmatrix} 0 & 1 \\ -1/2 & -1 \end{bmatrix} \begin{bmatrix} s_1 \\ s_2 \end{bmatrix}$$

Let's denote the system matrix as $A$:

$$A = \begin{bmatrix} 0 & 1 \\ -1/2 & -1 \end{bmatrix}$$

### Step 2: Check Eigenvalues of the System Matrix

We'll first verify that the system is stable by finding the eigenvalues of $A$:

The characteristic equation is:
$$\det(A - \lambda I) = 0$$

$$\det\begin{bmatrix} -\lambda & 1 \\ -1/2 & -1-\lambda \end{bmatrix} = 0$$

$$(-\lambda)(-1-\lambda) - 1 \cdot (-1/2) = 0$$

$$\lambda(1+\lambda) + 1/2 = 0$$

$$\lambda^2 + \lambda + 1/2 = 0$$

Using the quadratic formula:
$$\lambda = \frac{-1 \pm \sqrt{1 - 4 \cdot 1 \cdot 1/2}}{2} = \frac{-1 \pm \sqrt{1 - 2}}{2} = \frac{-1 \pm \sqrt{-1}}{2}$$

Thus, the eigenvalues are:
$$\lambda_1 = \frac{-1 + i}{2}$$
$$\lambda_2 = \frac{-1 - i}{2}$$

Since both eigenvalues have negative real parts, the system is asymptotically stable. Now we'll prove this using a Lyapunov function.

### Step 3: Construct a Lyapunov Function

For a Lyapunov function, we seek a positive definite matrix $P$ that satisfies the Lyapunov equation:

$$A^T P + P A = -Q$$

where $Q$ is a positive definite matrix.

It's important to note that this equation is directly related to the derivative of the Lyapunov function. Consider a quadratic Lyapunov function $V(s) = s^T P s$. Its derivative along the system trajectories is:

$$\dot{V}(s) = \dot{s}^T P s + s^T P \dot{s}$$

Since $\dot{s} = As$, we get:

$$\dot{V}(s) = (As)^T P s + s^T P (As) = s^T A^T P s + s^T P A s = s^T (A^T P + P A) s$$

Therefore, if $A^T P + P A = -Q$, then $\dot{V}(s) = -s^T Q s$. When $Q$ is positive definite, this ensures that the derivative of the Lyapunov function is negative definite, which is necessary for stability.

To obtain integer values for matrix $P$, instead of choosing $Q = I$, let's select $Q = \begin{bmatrix} 2 & 0 \\ 0 & 4 \end{bmatrix}$.

Let's find $P = \begin{bmatrix} p & q \\ q & r \end{bmatrix}$ that satisfies:

$$A^T P + P A = -Q = -\begin{bmatrix} 2 & 0 \\ 0 & 4 \end{bmatrix}$$

First, calculate $A^T$:

$$A^T = \begin{bmatrix} 0 & -1/2 \\ 1 & -1 \end{bmatrix}$$

Expanding the Lyapunov equation:

$$\begin{bmatrix} 0 & -1/2 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} p & q \\ q & r \end{bmatrix} + \begin{bmatrix} p & q \\ q & r \end{bmatrix} \begin{bmatrix} 0 & 1 \\ -1/2 & -1 \end{bmatrix} = -\begin{bmatrix} 2 & 0 \\ 0 & 4 \end{bmatrix}$$

Computing the left side:

$$\begin{bmatrix} -q/2 & -r/2 \\ p-q & q-r \end{bmatrix} + \begin{bmatrix} -q/2 & p-q \\ -r/2 & q-r \end{bmatrix} = \begin{bmatrix} -q & p-q-r/2 \\ p-q-r/2 & 2q-2r \end{bmatrix}$$

Equating with $-Q$:

$$\begin{bmatrix} -q & p-q-r/2 \\ p-q-r/2 & 2q-2r \end{bmatrix} = -\begin{bmatrix} 2 & 0 \\ 0 & 4 \end{bmatrix}$$

This gives us the following system of equations:
1. $-q = -2 \Rightarrow q = 2$
2. $p-q-r/2 = 0 \Rightarrow p = q + r/2 = 2 + r/2$
3. $2q-2r = -4 \Rightarrow 4-2r = -4 \Rightarrow r = 4$

Substituting $r = 4$ into equation 2:
$$p = 2 + 4/2 = 2 + 2 = 4$$

So, our solution is $p = 4$, $q = 2$, $r = 4$, giving us:

$$P = \begin{bmatrix} 4 & 2 \\ 2 & 4 \end{bmatrix}$$

### Step 4: Verify $P$ is Positive Definite

We need to ensure that $P$ is positive definite by checking if its eigenvalues are positive.

The characteristic equation for $P$ is:
$$\det(P - \lambda I) = 0$$

$$\det\begin{bmatrix} 4 - \lambda & 2 \\ 2 & 4 - \lambda \end{bmatrix} = 0$$

$$(4 - \lambda)(4 - \lambda) - 2 \cdot 2 = 0$$

$$(4 - \lambda)^2 - 4 = 0$$

$$(4 - \lambda)^2 = 4$$

$$4 - \lambda = \pm 2$$

$$\lambda = 4 \mp 2$$

Therefore, the eigenvalues are:
$$\lambda_1 = 2$$
$$\lambda_2 = 6$$

Since both eigenvalues are positive, $P$ is indeed positive definite.

### Step 5: Define the Lyapunov Function

The Lyapunov function for our system is:

$$V(s) = s^T P s = \begin{bmatrix} s_1 & s_2 \end{bmatrix} \begin{bmatrix} 4 & 2 \\ 2 & 4 \end{bmatrix} \begin{bmatrix} s_1 \\ s_2 \end{bmatrix}$$

Expanding:

$$V(s) = 4s_1^2 + 4s_1s_2 + 4s_2^2 = 4(s_1^2 + s_1s_2 + s_2^2)$$

### Step 6: Calculate the Derivative of the Lyapunov Function

The derivative of $V$ along the system trajectories is:

$$\dot{V}(s) = \dot{s}^T P s + s^T P \dot{s} = s^T (A^T P + P A) s$$

Since we constructed $P$ to satisfy $A^T P + P A = -Q$ where $Q = \begin{bmatrix} 2 & 0 \\ 0 & 4 \end{bmatrix}$, we have:

$$\dot{V}(s) = s^T (-Q) s = -s^T \begin{bmatrix} 2 & 0 \\ 0 & 4 \end{bmatrix} s = -2s_1^2 - 4s_2^2$$

### Step 7: Verify Stability Conditions

For a Lyapunov function to establish asymptotic stability, it must satisfy:

1. $V(s) > 0$ for all $s \neq 0$ (positive definite)
2. $\dot{V}(s) < 0$ for all $s \neq 0$ (negative definite)

We've shown:
- $V(s) = 4s_1^2 + 4s_1s_2 + 4s_2^2 = 4(s_1^2 + s_1s_2 + s_2^2)$ is positive definite (since $P$ is positive definite with eigenvalues 2 and 6)
- $\dot{V}(s) = -2s_1^2 - 4s_2^2$ is negative definite (since both coefficients are negative)

Therefore, by Lyapunov's direct method, the system is globally asymptotically stable.

## Conclusion

We have successfully demonstrated the stability of the given system with parameters $(k_1, k_2) = (1/2, -1)$ using a quadratic Lyapunov function:

$$V(s) = 4s_1^2 + 4s_1s_2 + 4s_2^2 = 4(s_1^2 + s_1s_2 + s_2^2)$$

This function satisfies all the conditions of Lyapunov's stability theorem, confirming that the system is globally asymptotically stable.

Note that we were able to find a Lyapunov function with integer coefficients by carefully choosing the matrix Q in the Lyapunov equation instead of using the identity matrix.

# Instability Proof for Dynamical System

## Problem Statement

Consider the dynamical system given by:

$$\begin{bmatrix} \dot{s}_1 \\ \dot{s}_2 \end{bmatrix} = \begin{bmatrix} 0 & 1 \\ -1+k_1 & k_2 \end{bmatrix} \begin{bmatrix} s_1 \\ s_2 \end{bmatrix}$$

with parameters $(k_1, k_2) = (2, 1)$.

We need to prove the instability of this system.

## Solution

### Step 1: Substitute the Parameter Values

First, let's substitute the parameter values into the system matrix:

### Step 3: Proof Using Lyapunov Theory

Lyapunov's theory provides powerful methods for analyzing the stability of dynamical systems without explicitly solving the differential equations. It involves constructing a scalar function of the state variables and examining its behavior along system trajectories.

#### Lyapunov Function Properties

A Lyapunov function $V(s)$ for stability analysis typically has the following properties:

1. **Continuity**: $V(s)$ should be continuous with continuous first partial derivatives.
2. **Local positive definiteness**: $V(0) = 0$ and $V(s) > 0$ for all $s \neq 0$ in some neighborhood of the origin.
3. **Derivative behavior**: The time derivative of $V(s)$ along system trajectories provides stability information.

For stability analysis:
- If $\dot{V}(s) \leq 0$ for all $s$ in some neighborhood of the origin, the system is stable.
- If $\dot{V}(s) < 0$ for all $s \neq 0$ in some neighborhood of the origin, the system is asymptotically stable.

For instability analysis:
- If there exists a continuous function $V(s)$ such that $V(0) = 0$ and in every neighborhood of the origin, there are points where $V(s) > 0$ and $\dot{V}(s) > 0$, then the system is unstable.

#### Chetaev's Instability Theorem

For proving instability, we often use Chetaev's theorem, which states:
If there exists a function $V(s)$ such that:
1. $V(0) = 0$
2. In every neighborhood of the origin, there are points where $V(s) > 0$
3. For these points, $\dot{V}(s) > 0$

Then the system is unstable.

Let's apply Chetaev's theorem to prove the instability of our system. Consider the function:

$V(s) = s_1^2$

This function satisfies $V(0) = 0$, and $V(s) > 0$ for all $s_1 \neq 0$.

Now, let's compute the derivative of $V$ along the system trajectories:

$\dot{V}(s) = \frac{d}{dt}(s_1^2) = 2s_1 \dot{s}_1$

From our system equations, we know that $\dot{s}_1 = s_2$, so:

$\dot{V}(s) = 2s_1 s_2$

For points where $s_1 > 0$ and $s_2 > 0$ (first quadrant), or where $s_1 < 0$ and $s_2 < 0$ (third quadrant), we have $\dot{V}(s) > 0$. This means that the function $V(s)$ is increasing along these trajectories, and the system's state moves away from the origin in these regions.

Therefore, according to Chetaev's theorem, the system is unstable. No matter how close to the origin we start, there exist initial conditions from which the system's trajectories will diverge.
